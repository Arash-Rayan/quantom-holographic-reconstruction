```python
from PIL import Image
import torchvision.transforms.functional as TF

img_path = 'transformed_o.png'
pred_angle = 45

# Read image (PIL object)
img = Image.open(img_path).convert("RGB")

# Rotate
aligned = TF.rotate(img, -pred_angle)

# Show
aligned.show()

# If you want to save
aligned.save("rotated.png")
```
